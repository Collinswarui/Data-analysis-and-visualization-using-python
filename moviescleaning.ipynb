```python
#import the necessary libraries

import pandas as pd 
import numpy as np
import seaborn as sns
import matplotlib
import matplotlib.pyplot as plt


```

```python
df=pd.read_csv('/home/anonymous/datasets/movies.csv')
```

```python
#print the first 50 rows

df.head(50)
```

```python
#insert a random column called No_ofViews

No_ofViews=np.random.randint(1000000,size=len(df))
df.insert(1,'No_ofViews',No_ofViews)
df
```

```python
#Get to know the description of the dataset

df.describe()
```

```python
#Renaming of colummns

df=df.rename(columns={"MOVIES":"Movies","YEAR":"Year","GENRE":"Genre","RATING":"Rating","ONE-LINE":"One_line","STARS":"Stars","VOTES":"Votes"})
```
